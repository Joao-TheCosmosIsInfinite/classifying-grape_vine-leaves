# Grape Vines Disease Classificator

## Imports/ Libs

```python
## Imports/ Libs
import pathlib
import numpy as np
from PIL import ImageTk, Image
```

## Files and Images

```python
## Instantiating the database in memory
url_database = '../database'
data_dir = pathlib.Path(url_database)
```

```python
## Counting the number of images
image_count = len(list(data_dir.glob('**/*.[jJ][pP][gG]')))
print(f"Image Totals: {image_count}")
```

```python
## Identifying the subfolders with the images
sub_folders = [f.name for f in data_dir.iterdir() if f.is_dir()]
sub_folders
```

### See some images/ files

```python
## List of path with images
leafblight = list(data_dir.glob('LeafBlight/*'))
blackRot = list(data_dir.glob('BlackRot/*'))
```

```python
## 1st leafblight image
Image.open(str(leafblight[0]))
```

```python
## 1st blackRot image
Image.open(str(blackRot[0]))
```

```python
## How many images are in each folder?
for subfolder in sub_folders:
    path = data_dir/ subfolder
    images = list(path.glob('*.jpg'))
    print(f'* Class {subfolder} has {len(images)} images')

    if images:
        fist_img = Image.open(str(images[0]))
        first_img_array = np.array(fist_img)

        last_img = Image.open(str(images[-1]))
        last_img_array = np.array(last_img)
        
        print(f'** Dimension of first image in {subfolder}:{first_img_array.shape}')
        print(f'** Dimension of last image in {subfolder}:{last_img_array.shape}')
```
